```python
# 라이브러리 로드
import cv2
import numpy as np
import time
```

#### 기본 출력

```python
# 웹캠디바이스 연결하는 데 시간소요
cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

while True:
    ret, img = cap.read()  # 웹캠 화면은 하나씩 로드, ret(리턴결과), img(실제 이미지)
    if not ret: break

    cv2.imshow('webcam', img)
    if cv2.waitKey(1) == ord('q'): break   # ord('A') = 65, ord('a' = 97) q(113)키를 누르면 반복 종료

cap.release() # 메모리 정리
cv2.destroyAllWindows()  
```

#### FPS(초당 프레임 수) 표시
- 대부분 30 FPS 사용함

```python
import time

cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

prev_time = 0     # 이전 프레임시간

while True:
    ret, img = cap.read()
    if not ret: break

    curr_time = time.time()  # 프레임 출력시간
    # FPS 계산
    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time 

    # 출력할 텍스트
    fps_text = f'FPS: {int(fps)}'  # int() 사용안하면 30.4534543233 소수점 출력됨
    # img 위에 텍스트 붙이기
    # cv2.putText(img=img, text=fps_text, org=(10, 30), 
    #             fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
    #             fontScale=0.9, color=(0, 255, 0), thickness=2)
    cv2.putText(img, fps_text, (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    cv2.imshow('webcam', img)
    if cv2.waitKey(1) == ord('q'): break

cap.release()
cv2.destroyAllWindows()
```

#### 얼굴 검출
- OpenCV 내장
    - haarcascade_frontalface_default.xml
    - haarcascade_eye_tree_eyeglasses.xml
    - haarcascade_eye.xml
    - haarcascade_smile.xml
    - haarcascade_frontalcatface_extended.xml

```python
# Haar Cascade 파일 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') # 얼굴데이터 따로
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')   # 미소데이터 따로

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패!')

real_w = cap.get(cv2.CAP_PROP_img_WIDTH)
real_h = cap.get(cv2.CAP_PROP_img_HEIGHT)

prev_time = 0
show_info = False

while True:
    ret, img = cap.read()
    if not ret: break

    # FPS
    curr_time = time.time()
    
    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time 

    if show_info:
        fps_text = f'FPS: {int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'
        cv2.putText(img, fps_text, (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        cv2.putText(img, res_text, (10, 70), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # image : 검출할 이미지
    # scaleFactor : 이미지를 몇 배씩 줄여가며 탐색할지 지정하는 값, 1.1 -> 10% 줄이기(속도저하, 정확도상승) 1.7 -> 70% 줄이기(속도빠름, 정확도저하)
    # 얼굴 1.1 ~ 1.3 (10~30%), 미소 : 1.5 ~1.8 
    # minNeighbors : 후보영역을 진짜 객체로 인정하려면 주변에서 몇 번이나 겹쳐 검출되어야하는지 횟수
    faces = face_cascade.detectMultiScale(image=img, scaleFactor=1.1, minNeighbors=8)
    for (x, y, w, h) in faces:  # x, y 좌측상단좌표, w, h 너비와 높이
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 255), 2)

        # 얼굴 영역만 ROI(Region of Interect)
        face_roi = img[y:y+h, x:x+w]  

        # 얼굴내부에서 미소 검출
        smiles = smile_cascade.detectMultiScale(face_roi, 1.7, 20)

        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(face_roi, (sx, sy), (sx+sw, sy+sh), (0, 255, 0), 2)
        

    cv2.imshow('webcam', img)

    key = cv2.waitKey(1)
    if key == ord('q'): 
        break
    elif key == ord('s'):
        filename = time.strftime('snapshot_%y%m%d%H%M%S.jpg')
        cv2.imwrite(filename, img) 
        # print(f'{filename} 스냅샷 저장')
    elif key == 9:
        show_info = not show_info
        # print(f'표시여부 {show_info}')

cap.release()
cv2.destroyAllWindows()
```

#### 얼굴 모자이크

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') # 얼굴데이터 따로

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패!')

real_w = cap.get(cv2.CAP_PROP_img_WIDTH)
real_h = cap.get(cv2.CAP_PROP_img_HEIGHT)

prev_time = 0
show_info = False

while True:
    ret, img = cap.read()
    if not ret: break

    # FPS
    curr_time = time.time()
    
    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time 

    if show_info:
        fps_text = f'FPS: {int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'
        cv2.putText(img, fps_text, (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        cv2.putText(img, res_text, (10, 70), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    faces = face_cascade.detectMultiScale(image=img, scaleFactor=1.1, minNeighbors=8)
    for (x, y, w, h) in faces:  # x, y 좌측상단좌표, w, h 너비와 높이
        # cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 255), 2)

        # 얼굴 영역만 ROI(Region of Interect)
        face_roi = img[y:y+h, x:x+w]

        # 모자이크 처리
        small = cv2.resize(face_roi, (w//15, h//15), interpolation=cv2.INTER_LINEAR) # 작게 줄이고
        mosaic = cv2.resize(small, (w, h), interpolation=cv2.INTER_NEAREST)  # 다시 크게 늘리기 픽셀뭉개기

        img[y:y+h, x:x+w] = mosaic  # ROI로 교체

    cv2.imshow('webcam', img)

    key = cv2.waitKey(1)
    if key == ord('q'): 
        break
    elif key == ord('s'):
        filename = time.strftime('snapshot_%y%m%d%H%M%S.jpg')
        cv2.imwrite(filename, img) 
    elif key == 9:
        show_info = not show_info

cap.release()
cv2.destroyAllWindows()
```

#### QR코드 리더

- pyzbar 라이브러리 - 구식 라이브러리로 호환 문제 발생, zbar 기초 라이브러리 문제 발생가능

```python
# 사용안함
import pyzbar.pyzbar as pyzbar  # !pip install pyzbar

# QR코드 검출기 객체생성
qr_reader = cv2.QRCodeDetector()
# 웹캠 오픈
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    if not ret: break

    # 회색톤으로
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # pyzbar 
    decoded = pyzbar.decode(gray)
    barcode_data = ''

    for d in decoded:
        x, y, w, h = d.rect

        barcode_data = d.data.decode('utf-8')
        barcode_type = d.type

        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

        text = f'{barcode_data} ({barcode_type})'
        cv2.putText(img, text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('QR Code Reader', img)

    key = cv2.waitKey(1)
    if key == ord('q'): break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import webbrowser  # 웹브라우저용 모듈 로드

qr_detector = cv2.QRCodeDetector()  # QR 코드 검출기 생성
cap = cv2.VideoCapture(0)       # 웹캠 열기
is_opened = False  # 웹사이트 오픈 여부
last_data = ''  # 최종데이터

while True:
    ret, img = cap.read()
    if not ret: break

    data, points, _ = qr_detector.detectAndDecode(img)

    if points is not None and data:
        points = points.astype(int)

        for i in range(len(points[0])):  # QR코드 외곽포인트 네개
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i + 1) % len(points[0])])
            cv2.line(img, pt1, pt2, (0, 255, 0), 2)

        cv2.putText(img, data, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # 같은 QR 반복실행 방지
        if (data.startswith('http://') or data.startswith('https://')) and (not is_opened and last_data != data):
            webbrowser.open(data)
            is_opened = True; last_data = data
        else:
            is_opened = False

    cv2.imshow("QR Code Reader", img)
    if cv2.waitKey(1) == ord('q'): break    

cap.release()
cv2.destroyAllWindows()
```
