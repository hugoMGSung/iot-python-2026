#### 기본 사용법

````python
# OpenCV 라이브러리 로드, 넘파이가 필요할 수 있음
import cv2
import numpy as np
````

````python
# 버전확인
cv2.__version__
````

````python
np.__version__
````

````python
# 이미지로드
img = cv2.imread('./cat01.jpg')

cv2.imshow('origin', img)   # origin 윈도우 제목으로 이미지 오픈
cv2.waitKey(0)      # 키보드 입력 대기, 무슨키든지 다 종료
cv2.destroyAllWindows()  # 종료 후 열린 모든 윈도우창 닫음
````

````python
# 동영상로드
cap = cv2.VideoCapture(filename='./sample_360p.mp4')

while True:  # 무한루프
    ret, frame = cap.read()  # FPS의 이미지 하나씩 읽어옴
    if not ret: break  # 더이상 읽을 이미지가 없으면 종료

    cv2.imshow('video', frame) # 이미지를 계속 변경
    if cv2.waitKey(1) == ord('q'): break   # 키보드 q입력하면 종료

cap.release()  # 메모리 해제
cv2.destroyAllWindows()
````

````python
# 실시간 웹캠 
cap = cv2.VideoCapture(0) # 웹캠 디바이스 번호 

while True: 
    ret, frame = cap.read() 
    if not ret: break 

    cv2.imshow('webcam', frame)
    if cv2.waitKey(1) == ord('q'): break

cap.release() 
cv2.destroyAllWindows()
````
