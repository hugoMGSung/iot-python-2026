#### 위경도 특성
- 소수점 4자리이상 사용해야 제대로된 위치를 잡을 수 있음

````python
# folium 라이브러리 로드
import folium

# 37.551192, 126.988165(N서울타워)
````

````python
loc = [35.117491, 129.090457]

m = folium.Map(location=loc, zoom_start=20, tiles='CartoDB positron')  # 지도 객체 생성, m 변수 할당

# 마커를 생성해서 지도변수에 추가
folium.Marker(location=loc, 
              tooltip='부산', 
              popup='용당7관',
              icon=folium.Icon(icon='car', prefix='fa')).add_to(m)

m
````
